- Using a `for` loop, iterate over the number range 0-20. Within the loop you should:
    - Create a variable `station` that contains 1) the text from the `basename` variable, 2) the number, and 3) the file extension `.txt`  
    - Add the content of `station` to the `filenames` list, which should have following content in the end:

      ```

       'station_4.txt', 'station_5.txt', 'station_6.txt', 'station_7.txt',
      ```

````python
# YOUR CODE HERE
raise NotImplementedError()
````

````python
# using for loop and range function
for station in range(0,21):
# creating a variable named station that contains basename variable, number, and .txt using f-string and changing station number as string text
    station = f"{basename}_{str(station)}.txt"
#the content of the station is added inside the filenames list
    filenames.append(station)
````

````python
# Test print of the results
print(filenames)
````
