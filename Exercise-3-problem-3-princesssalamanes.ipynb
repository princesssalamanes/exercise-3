### Part 3 (0 points)

- Make a loop that iterates `n` times, and allocates stations to different geographical zones based on their coordinates. 

```python
# YOUR CODE HERE
raise NotImplementedError()
```

```python
# Using for loop and conditional statements to classify station based on their geographical zone.
for n in range(len(stations)):
    if lats[n] > north_south_cutoff and lons[n] < east_west_cutoff:
        north_west.append(stations[n])
    elif lats[n] > north_south_cutoff and lons[n] > east_west_cutoff:
        north_east.append(stations[n])
    elif lats[n] < north_south_cutoff and lons[n] < east_west_cutoff:
        south_west.append(stations[n])
    else:
        south_east.append(stations[n])
```
