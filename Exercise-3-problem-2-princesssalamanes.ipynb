### Part 2 (2 points)

Iterate over the temperatures and add temperatures to the different temperature classes defined below:

- **Cold**: Temperatures of less than -2 degrees
- **Slippery**: Temperatures equal to or warmer than -2 degrees, but less than +2 degrees
- **Comfortable**: Temperatures equal to or warmer than +2 degrees, but less than +15 degrees
- **Warm**: Temperatures equal to or warmer than +15 degrees

```python
# YOUR CODE HERE
raise NotImplementedError()
```

```python
# Temperature is classify into different temperature classes
for temperature in temperatures:
    if temperature < -2:
# Temperatures below -2 degrees will be added to the cold list
        cold.append(temperature)
    elif (temperature >=-2 and temperature < 2):
# Temperatures equal to or warmer than -2 degrees, but less than +2 degrees will be added to the slippery list
        slippery.append(temperature)
    elif (temperature >= 2 and temperature < 15):
# Temperatures equal to or warmer than +2 degrees, but less than +15 degrees
        comfortable.append(temperature)
    elif temperature >15:
# Temperatures equal to or warmer than +15 degrees
        warm.append(temperature)
```

```python
# Test prints for all lists (now they should contain values)
print(cold, slippery, comfortable, warm)
```
